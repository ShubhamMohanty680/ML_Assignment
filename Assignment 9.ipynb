# Q.13

There are only two types of apps:
      <ol type='1'>Free</ol>
     <ol>Paid</ol>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Change the Price data to float value
df['Price'] = df['Price'].str.split('$').str[1].astype(float)

# Group the dataset by app type and calculate the average price
grouped = df.groupby('Type')['Price'].mean()
grouped.fillna(0, inplace=True)

# Create a bar chart
plt.bar(grouped.index, grouped.values)
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')

# Show the plot
plt.show()
```

# Q.15

To find the most popular app categories:

   - My approach would be make a table with columns 'sum up the total no of installs by category'
   - Then sort the table by the number of installs in descending order.
   - This will give the list of most popular app categories.

```python
freq_table = pd.DataFrame(df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False).reset_index())
freq_table
```

```python
freq_table.boxplot(column='Installs', by='Category',
                   figsize=(20, 12), color='blue', rot=90)
plt.xlabel('Category')
plt.ylabel('Installs')
plt.title('Comparison of total installs by Category')

plt.show()
```
